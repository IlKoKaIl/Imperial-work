#### 1.1 Neutron Time of Flight

The energy of the neutrons has a small Gaussian spread (from doppler broadening), thus the arrivals times will also be Gaussian-distributed. But they will be distorted the by the detector's own Gaussian timing response.

So the plan is:
Energy distribution -> time of flight distribution -> measured signal -> our param estimation.

(i) The  saddlepoint  approximation  finds  a  Gaussian  approximation  of  a  strongly  peaked  function  of  the form exp [−f (x )] by expanding about the exponent function about its maxima:

$g(x) \approx \exp \left[-f\left(x_0\right)\right] \exp \left[-\frac{1}{2} f^{\prime \prime}\left(x_0\right)\left(x-x_0\right)^2\right]$

(a)Use this method to find an expression for the TOF distribution of deuterium fusion neutrons at the detector,\
$g = \frac{dN}{dt}$ using $f = -ln \frac{dN}{dE}$

so we have the intrinsic neutron energy distribution as
$$ \frac{dN}{dE} = \phi(E; E_{0}, \sigma_{E}) $$
and by subbing in velocity term into the energy equation we get this for the energy as a function of time:
$$
E(t) = \frac{1}{2} m_{n} \left( \frac{d}{t} \right)^2 = \frac{m_{n} d^2}{2t^2}
$$


So we want to use the saddlepoint approximation with $f = -ln \frac{dN}{dE}$ to get a Gaussian approximation for the true TOF distribution

So lets start with the definition of a Gaussian pdf 
$$\phi(X ; \mu, \sigma)=\frac{1}{\sqrt{2 \pi} \sigma} \exp \left[-\frac{(X-\mu)^2}{2 \sigma^2}\right] $$

So using what we were given we can do this (just subbing in terms): \
$ \frac{dN}{dE}(E) =  \frac{1}{\sqrt{2 \pi} \sigma} \exp \left[-\frac{(E-E_{0})^2}{2 \sigma_{E}^2}\right] $

Now if we evaluate this at $E = E(t) = \frac{m_{n} d^2}{2t^2}$ \
$ =>  \frac{1}{\sqrt{2 \pi} \sigma} \exp \left[-\frac{(E(t)-E_{0})^2}{2 \sigma_{E}^2}\right] $

We define the function \( f(t) \) according to the saddlepoint prescription:

$
f(t) = -\ln\!\left(\frac{dN}{dE}\Big|_{t}\right)
$
Taking the negative logarithm:

$= -\ln\!\left(\frac{1}{\sqrt{2\pi}\,\sigma_E}\right) + \frac{(E(t) - E_0)^2}{2\sigma_E^2}$
$= \frac{(E(t) - E_0)^2}{2\sigma_E^2} + \text{const.}$


### Step B — Find the peak \( t_0 \) from \( f'(t_0) = 0 \)

Differentiate:

$f'(t)= \frac{E(t) - E_0}{\sigma_E^2}\,E'(t)$

Setting $( f'(t_0) = 0 )$, and because $( E'(t_0) != 0 )$ we get

$
E(t_0) = E_0
$

Substitute the relation $( E(t) = \frac{m_n d^2}{2t^2})$:

$
E_0 = \frac{m_n d^2}{2t_0^2}
\quad\Rightarrow\quad
\boxed{
t_0 = d\sqrt{\frac{m_n}{2E_0}}
}
$

---

### Step C — Compute the curvature \( f''(t_0) \) and getting the variance from it

Differentiate again:

$f''(t)= \frac{[E'(t)]^2}{\sigma_E^2}+ \frac{E(t) - E_0}{\sigma_E^2}E''(t)$

At the peak $( E(t_0) = E_0)$, the second term vanishes, so

$
f''(t_0)
= \frac{[E'(t_0)]^2}{\sigma_E^2}
$

Now, using $( E(t) = \frac{m_n d^2}{2}t^{-2})$:

$
E'(t) = -\frac{m_n d^2}{t^3},
\qquad
E''(t) = \frac{3m_n d^2}{t^4}
$

Therefore,

$
[E'(t_0)]^2
= \frac{m_n^2 d^4}{t_0^6}
\quad\Rightarrow\quad
f''(t_0)
= \frac{m_n^2 d^4}{\sigma_E^2\,t_0^6}
$

---

### Step D — Variance of the TOF distribution

By the saddlepoint formula, the variance of the Gaussian approximation is the reciprocal of the curvature:

$
\sigma_{t,\text{true}}^2
= \frac{1}{f''(t_0)}
= \frac{\sigma_E^2\,t_0^6}{m_n^2 d^4}
$

Substitute $( t_0 = d\sqrt{\tfrac{m_n}{2E_0}})$:

$t_0^6= d^6\!\left(\frac{m_n}{2E_0}\right)^3$

Hence,

$\boxed{\sigma_{t,\text{true}}^2= \frac{\sigma_E^2\,d^2\,m_n}{8\,E_0^{3}}}$

---

### Step E — Gaussian approximation for the true TOF distribution

Finally, the time-of-flight distribution $(g(t) = \frac{dN}{dt})$ is approximately Gaussian with mean $(t_0)$ and variance $(\sigma_{t,\text{true}}^2)$:

$
\boxed{
g(t) \approx
\phi\!\Big(
t;\;
t_0 = d\sqrt{\tfrac{m_n}{2E_0}},\;
\sigma_{t,\text{true}} =
\sqrt{\tfrac{m_n d^2}{8E_0^{3}}}\,\sigma_E
\Big)
}
$
